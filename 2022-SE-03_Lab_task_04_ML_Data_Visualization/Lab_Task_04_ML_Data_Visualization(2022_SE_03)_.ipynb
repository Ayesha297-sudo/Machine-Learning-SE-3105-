# **Data Cleaning & Preprocessing**

```python
# Handling missing values
df.dropna(subset=['Cuisines'], inplace=True)  # Dropping rows with missing cuisines

# Convert categorical columns to binary (Yes -> 1, No -> 0)
binary_columns = ['Has Table booking', 'Has Online delivery', 'Is delivering now', 'Switch to order menu']
df[binary_columns] = df[binary_columns].applymap(lambda x: 1 if x == 'Yes' else 0)

# Dropping unnecessary columns (Address, Locality Verbose, etc.)
columns_to_drop = ['Address', 'Locality Verbose', 'Switch to order menu']
df.drop(columns=columns_to_drop, inplace=True)
```

#Key Insight
1. **Missing Data Handled:** Rows with missing cuisines were removed.
2. **Binary Conversion:** Columns like "Has Table booking" and "Has Online delivery" are now 0s and 1s instead of "Yes" and "No".
3. **Simplified Data:** Unnecessary columns like Address and Locality Verbose were dropped.
4. **Data types confirmed** (all columns correctly formatted).

#**Data Visualizations & Insights**

```python
# Set style for better visualization
sns.set_style("whitegrid")

# 1. Distribution of Aggregate Ratings
plt.figure(figsize=(8,5))
sns.histplot(df['Aggregate rating'], bins=10, kde=True, color="blue")
plt.title("Distribution of Restaurant Ratings")
plt.xlabel("Aggregate Rating")
plt.ylabel("Frequency")
plt.show()
```

# **1. Key insight from Restaurant Ratings Distribution**
1. Most restaurants have an aggregate rating between 2.5 and 4.5.
2. Very few restaurants have ratings below 2.0 or above 4.5.
3. The peak frequency is around 3.5 to 4.0, indicating that most restaurants receive moderate to high ratings.

```python
# 2. Top Cuisines by Count
plt.figure(figsize=(12,6))
top_cuisines = df['Cuisines'].value_counts().head(10)  # Top 10 most common cuisines
sns.barplot(x=top_cuisines.values, y=top_cuisines.index, palette="viridis")
plt.title("Top 10 Most Popular Cuisines")
plt.xlabel("Number of Restaurants")
plt.ylabel("Cuisines")
plt.show()
```

#**2. Key insight from Top Cuisines Chart**
* The most common cuisines are North Indian, Chinese, Fast
Food, Italian, and South Indian.
* North Indian cuisine is the most dominant.
* Fusion and global cuisines (e.g., Continental, Asian) are also widely available.

```python
# 3. Average Cost for Two vs. Aggregate Rating
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['Average Cost for two'], y=df['Aggregate rating'], alpha=0.5, color='red')
plt.title("Cost for Two vs. Aggregate Rating")
plt.xlabel("Average Cost for Two")
plt.ylabel("Aggregate Rating")
plt.xlim(0, df['Average Cost for two'].quantile(0.95))  # Limiting outliers for better visualization
plt.show()
```

# **3. Key insight from Cost vs. Ratings Scatter Plot**
* There is no strong correlation between cost and ratings.
* Some expensive restaurants (high cost for two) have high ratings, but many mid-range and low-cost restaurants also receive good ratings.
* Restaurants with moderate pricing (under 1000 currency units) tend to have a wider spread of ratings.

```python
# 4. Online Delivery vs. Ratings
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Has Online delivery'], y=df['Aggregate rating'], palette="coolwarm")
plt.xticks([0,1], ["No Online Delivery", "Online Delivery"])
plt.title("Online Delivery vs. Ratings")
plt.xlabel("Online Delivery Availability")
plt.ylabel("Aggregate Rating")
plt.show()
```

# **4. Key insight from Online Delivery vs. Ratings**
* Restaurants that offer online delivery tend to have slightly lower ratings than those without online delivery.
* This might indicate that dine-in restaurants focus more on quality service, while delivery restaurants handle large volumes with variable service quality.

```python
# 5. Table Booking vs. Ratings
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Has Table booking'], y=df['Aggregate rating'], palette="mako")
plt.xticks([0,1], ["No Table Booking", "Table Booking Available"])
plt.title("Table Booking vs. Ratings")
plt.xlabel("Table Booking Availability")
plt.ylabel("Aggregate Rating")
plt.show()
```

# **5. Key insight from Table Booking vs. Ratings**
* Restaurants that offer table booking tend to have higher ratings.
* This suggests that higher-end or well-managed restaurants provide better service and experience, leading to better customer ratings.

```python
# 6. City-wise Restaurant Distribution
plt.figure(figsize=(12,6))
top_cities = df['City'].value_counts().head(10)  # Top 10 cities with the most restaurants
sns.barplot(x=top_cities.values, y=top_cities.index, palette="Blues_r")
plt.title("Top 10 Cities with the Most Restaurants")
plt.xlabel("Number of Restaurants")
plt.ylabel("City")
plt.show()
```

# **6. Key insight from City-wise Restaurant Distribution**
* The highest number of restaurants are concentrated in a few key cities.
* Some cities have a much higher restaurant density, which could indicate a strong food culture or high urban population.
* This distribution can help identify competitive and emerging markets for restaurants.

```python
# 7. Price Range vs. Ratings
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Price range'], y=df['Aggregate rating'], palette="Set2")
plt.title("Price Range vs. Aggregate Rating")
plt.xlabel("Price Range (1 = Low, 4 = High)")
plt.ylabel("Aggregate Rating")
plt.show()
```

# **7. Key insights from Price Range vs. Ratings**
* Higher price range restaurants generally have higher ratings.
* Budget restaurants (Price Range = 1) have more variability, with both very low and some high ratings.
* Premium restaurants (Price Range = 4) tend to maintain a higher standard, leading to consistently better ratings.

```python
# 8. Votes Distribution
plt.figure(figsize=(10,5))
sns.histplot(df['Votes'], bins=30, kde=True, color="purple")
plt.title("Distribution of Votes for Restaurants")
plt.xlabel("Number of Votes")
plt.ylabel("Frequency")
plt.xlim(0, df['Votes'].quantile(0.95))  # Limiting extreme outliers
plt.show()
```

# **8. Key insight from Distribution of Votes for Restaurants**
**Votes distribution** is highly skewed, with some restaurants receiving a significantly higher number of reviews.

```python
# Identify the top restaurant chains by counting occurrences of each restaurant name
top_restaurant_chains = df['Restaurant Name'].value_counts().head(10)

# Plot the top restaurant chains by number of outlets
plt.figure(figsize=(12, 6))
sns.barplot(x=top_restaurant_chains.values, y=top_restaurant_chains.index, palette="viridis")
plt.xlabel("Number of Outlets")
plt.ylabel("Restaurant Chain")
plt.title("Top 10 Restaurant Chains by Number of Outlets")
plt.show()
```

# **9. Key insight from Top Restaurant Chains by Number of Outlets**

* The most popular restaurant chains have multiple outlets, with some dominating in specific regions.
* The top chains indicate brand loyalty and high consumer preference.

```python
numeric_df = df[['Average Cost for two', 'Aggregate rating', 'Votes', 'Price range']]
sns.heatmap(numeric_df.corr(), annot=True)
plt.title("Correlation Heatmap")
```

# **10. Key insight from Correlation Heatmap**
* Cost for two has minimal impact on ratings, votes, or price range.
* Ratings & votes show a moderate correlation (0.31), meaning popular restaurants often have higher ratings.
* Price range & ratings (0.44) suggest expensive restaurants tend to be rated better.
* Price range & votes (0.31) indicate premium restaurants attract more reviews.
* Affordability alone does not determine restaurant popularity.
